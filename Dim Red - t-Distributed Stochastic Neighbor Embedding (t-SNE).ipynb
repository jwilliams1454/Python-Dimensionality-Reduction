```python
# t-SNE
# https://cmdlinetips.com/2019/07/dimensionality-reduction-with-tsne/


#load packages
import matplotlib.pyplot as plt
import seaborn as sns
%matplotlib inline
import pandas as pd
```

```python
#load MNIST data set for pattern recognition of numbers
from sklearn.datasets import load_digits
digits = load_digits()
```

```python
# matrix size
digits.data.shape
```

```python
digits.target
```

```python
#  subset both the data set and the actual digit it correspond to.
data_X = digits.data[:600]
y = digits.target[:600]
```

```python
# load TSNE
from sklearn.manifold import TSNE
tsne = TSNE(n_components=2, random_state=0)
```

```python
# load dataset
tsne_obj= tsne.fit_transform(data_X)
tsne_obj
```

```python
# scatter plot to visualize the low-dimensional representation of the data

# store results from tSNE as a Pandas dataframe with the target integer for each data point
tsne_df = pd.DataFrame({'X':tsne_obj[:,0],
                        'Y':tsne_obj[:,1],
                        'digit':y})
tsne_df.head()
```

```python
# scatter plot with using the two arrays we got from tSNE
sns.scatterplot(x="X", y="Y",
              data=tsne_df);
```

```python
# color and label each digit.
sns.scatterplot(x="X", y="Y",
              hue="digit",
              palette=['purple','red','orange','brown','blue',
                       'dodgerblue','green','lightgreen','darkcyan', 'black'],
              legend='full',
              data=tsne_df);
```
